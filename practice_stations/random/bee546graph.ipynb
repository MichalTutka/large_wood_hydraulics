```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Specify the path to your CSV file
csv_file_path = r'D:\Michal\Documents\0_Grad_School_OSU\Classes\BEE 546 - River Engineering\max velocities.csv'

# Read the CSV file into a pandas DataFrame
data = pd.read_csv(csv_file_path)
```

```python
# Read the CSV file into a pandas DataFrame
data = pd.read_csv(csv_file_path)

# Assign each column to a variable
station = data.iloc[:, 0]
dd1984 = data.iloc[:, 1]
dd1976 = data.iloc[:, 2]
dd2001 = data.iloc[:, 3]

plt.figure(figsize=(16, 10))
plt.ylim(0,10)

# Plot columns 2-4 over column 1
plt.plot(station, dd1976, label='DD 1976', alpha=0.9)
plt.plot(station, dd2001, label='DD 2001', alpha=0.9)
plt.plot(station, dd1984, label='DD 1984', alpha=0.9)


# Set labels and title for the plot
plt.xlabel('Station (m)')
plt.ylabel('Max velocity (m/s)')
plt.title('Max velocity VS distance downstream')

# Show a legend
plt.legend()

# Display the plot
plt.show()

save_path = r'D:\Michal\Documents\0_Grad_School_OSU\Classes\BEE 546 - River Engineering\myfigure.png'
plt.savefig(save_path, dpi=300)
```

```python

combined_data = [dd1984, dd1976, dd2001]

# Set the number of bins for the histogram
num_bins = 100

# Plot the combined histogram
plt.figure(figsize=(16, 10))
plt.hist(combined_data, bins=num_bins, label=['dd 1984', 'dd 1976', 'dd 2001'])
plt.xlim(0,5)


# Set labels and title for the plot
plt.xlabel('Values (m/s)')
plt.ylabel('Frequency')
plt.title('Histogram of Max Velocities')

# Show a legend
plt.legend()

# Display the plot
plt.show()
```
